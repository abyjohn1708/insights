```python
import numpy as np
import seaborn as sns 
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('C:/Users/User/fifa_data.csv')
df
```

Exercise 1:
Using the below dataset of fifa players dataset. Perform Exploratory data analysis and find the following insights:
1.Which country has the most number of players

```python

most_no_of_players = df['Nationality'].value_counts()
print ("most number of players",most_no_of_players)
```

2.Plot a bar chart of 5 top countries with most number of players

```python
top_countries = df['Nationality'].value_counts().head(5)
plt.figure(figsize=(10, 6))
sns.barplot(x=top_countries.index, y=top_countries.values)
plt.title('Top 5 Countries with Most Players')
plt.xlabel('Country')
plt.ylabel('Number of Players')
plt.show()

```

3.Which player has the highest salary?

```python
df['Wage'] = df['Wage'].astype(str)  # Convert the 'Wage' column to string type
df['Wage'] = df['Wage'].str.replace('€', '')  # Remove the '€' symbol if present
df['Wage'] = df['Wage'].str.replace('K', '000')  # Replace 'K' with '000' to convert thousands
df['Wage'] = df['Wage'].str.replace('M', '000000')  # Replace 'M' with '000000' to convert millions
df['Wage'] = pd.to_numeric(df['Wage'])  # Convert the 'Wage' column to numeric data type

highest_salary = df['Wage'].max()
highest_salary_player = df.loc[df['Wage'].idxmax(), 'Name']
print("Player with the highest salary:", highest_salary_player)
print("The highest salary:", highest_salary)
```

4.Plot a histogram to get the salary range of the players.

```python
plt.figure(figsize=(10, 3))
sns.histplot(df['Wage'], bins=20)
plt.title('Salary Range of Players')
plt.xlabel('Salary')
plt.ylabel('Number of players')
plt.show() 
```

5. Who is the tallest player in the FIFA players?

```python
df['Height'] = df['Height'].astype(str)  # Convert the 'Height' column to string type
df['Height'] = df['Height'].str.replace("'", ".")  # Replace the single quote with a dot
df['Height'] = pd.to_numeric(df['Height'])  # Convert the 'Height' column to numeric type

tallest_player = df.loc[df['Height'].idxmax(), 'Name']
tallest_height=df['Height'].max()
print("The tallest player is:", tallest_player)
print(f"Tallest player :{tallest_height} inchs")
```

6.Which club has most number of players?

```python
club_with_most_players = df['Club'].value_counts().idxmax()
print("Club with the most number of players:", club_with_most_players)
```

7.Which foot is most preferred by the players?Draw a bar chart for preferred foot

```python
preferred_foot_counts = df['Preferred Foot'].value_counts()
plt.figure(figsize=(4, 3))
sns.barplot(x=preferred_foot_counts.index, y=preferred_foot_counts.values)
plt.title('Preferred Foot of Players')
plt.xlabel('Preferred Foot')
plt.ylabel('Number of Players')
plt.show()
```
